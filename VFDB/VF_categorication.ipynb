```python
import pandas as pd
from Bio import SeqIO
import re
import json
import os
```

```python
setA_fasta = SeqIO.to_dict(SeqIO.parse("VFDB_setA_nt.fas", 
                                       "fasta"
                                      )
                          )
```

# vf_cate

```python
vfinfo = pd.read_excel("VFs.xls",
                       sheet_name = 'Sheet1'
#                      sep = "\t"
                    )
vfinfo = vfinfo.set_index("VFID")
vfinfo
```

# vf info

```python
seqID_VFID = {}

for keys in setA_fasta.keys():
    seqID_VFID[keys] = re.findall("VF[0-9][0-9][0-9][0-9]", setA_fasta[keys].description)
```

```python
seqID_VFID
```

```python
def find(string, char):
    for i, c in enumerate(string):
        if c == char:
            yield i
```

```python
for x in seqID_VFID.keys():
    ind_1 = list(find(setA_fasta[x].description, "["))[-1]
    ind_2 = list(find(setA_fasta[x].description, "("))[1]
    ind_3 = list(find(setA_fasta[x].description, ")"))[1]
    seqID_VFID[x].append(setA_fasta[x].description[ind_1 + 1: -1])
    seqID_VFID[x].append(vfinfo.loc[seqID_VFID[x][0], "VF_Name"])
    seqID_VFID[x].append(vfinfo.loc[seqID_VFID[x][0], "VFcategory"])
    seqID_VFID[x].append(setA_fasta[x].description[ind_2 + 1: ind_3])
```

```python
seqID_VFID
```

```python
vfinfo_customized_df = pd.DataFrame.from_dict(seqID_VFID, 
                                              orient = "index")
vfinfo_customized_df.columns = ["VFID", "strain", "VF_Name", "cate", "pro_Name"]
vfinfo_customized_df
```

```python
vfinfo_customized_df.to_csv("processed_vfinfo.txt")
```
